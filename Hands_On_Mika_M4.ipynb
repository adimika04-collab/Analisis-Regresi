```python
# ===============================================
# TUGAS 2 - ANALISIS KORELASI LENGKAP
# Dataset: Pengalaman Kerja dan Gaji
# ===============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# ===============================================
# 1. Memuat Dataset Eksternal (.csv)
# ===============================================
data = pd.read_csv("/content/Data_Pengalaman_Kerja_dan_Gaji_1000.csv")

# Memilih variabel
X = data["Pengalaman_Kerja_Tahun"]
Y = data["Gaji_Juta"]

print("Jumlah Data:", len(data))
print("\n5 Data Pertama:")
print(data.head())

# ===============================================
# 2. Visualisasi Scatterplot + Garis Regresi
# ===============================================
plt.figure(figsize=(8,6))

# Scatterplot
plt.scatter(X, Y)

# Hitung garis regresi
m, b = np.polyfit(X, Y, 1)
plt.plot(X, m*X + b)

plt.xlabel("Pengalaman Kerja (Tahun)")
plt.ylabel("Gaji (Juta Rupiah)")
plt.title("Scatterplot dan Garis Regresi")
plt.grid(True)
plt.show()

print("\nPersamaan Regresi:")
print(f"Y = {round(m,3)}X + {round(b,3)}")

# ===============================================
# 3. Menghitung Korelasi Pearson
# ===============================================
r, p_value = pearsonr(X, Y)

print("\nKoefisien Korelasi Pearson (r):", round(r,3))
print("P-value:", round(p_value,5))

# ===============================================
# 4. Uji Hipotesis
# ===============================================
alpha = 0.05

print("\nHipotesis:")
print("H0: Tidak ada hubungan linier antara pengalaman kerja dan gaji")
print("H1: Ada hubungan linier antara pengalaman kerja dan gaji")

if p_value < alpha:
    keputusan = "Tolak H0"
    kesimpulan = "Terdapat hubungan linier yang signifikan."
else:
    keputusan = "Gagal Tolak H0"
    kesimpulan = "Tidak terdapat hubungan linier yang signifikan."

print("\nKeputusan:", keputusan)
print("Kesimpulan:", kesimpulan)

# ===============================================
# 5. Interpretasi Kekuatan Korelasi
# ===============================================
if r > 0.8:
    interpretasi = "Korelasi sangat kuat positif"
elif r > 0.6:
    interpretasi = "Korelasi kuat positif"
elif r > 0.4:
    interpretasi = "Korelasi sedang positif"
elif r > 0.2:
    interpretasi = "Korelasi lemah positif"
elif r > 0:
    interpretasi = "Korelasi sangat lemah positif"
elif r < -0.8:
    interpretasi = "Korelasi sangat kuat negatif"
elif r < -0.6:
    interpretasi = "Korelasi kuat negatif"
elif r < -0.4:
    interpretasi = "Korelasi sedang negatif"
elif r < -0.2:
    interpretasi = "Korelasi lemah negatif"
elif r < 0:
    interpretasi = "Korelasi sangat lemah negatif"
else:
    interpretasi = "Tidak ada korelasi linier"

print("\nInterpretasi Kekuatan Korelasi:", interpretasi)

# ===============================================
# 6. Koefisien Determinasi (R²)
# ===============================================
R2 = r**2
print("Koefisien Determinasi (R²):", round(R2,3))
print("Artinya sekitar", round(R2*100,1), "% variasi gaji dijelaskan oleh pengalaman kerja.")
```
